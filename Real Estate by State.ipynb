```python
import pandas as pd
import os
import re
import datetime
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('real_estate_state.csv')
#create a new column to format the date as a date
df['date'] = pd.to_datetime(df['month_date_yyyymm'], format= '%Y%m')
```

```python
#create new dataframe specific by state
state = 'texas'
st_df = df.loc[df['state'].str.contains(state)].reset_index()
#add two columns to split the date
st_df['year'] = [date.year for date in st_df.date]
st_df['month'] = [month.month for month in st_df.date]
st_df = st_df.drop(columns = ['index', 'month_date_yyyymm'])

lst = []
for column_name in st_df.columns:
    if '_mm' in column_name:
        lst.append(column_name)
    elif '_yy' in column_name:
        lst.append(column_name)
st_df = st_df.drop(columns = lst)
st_df
```

```python
x = st_df.month
y = st_df.median_listing_price

plt.plot(x, y)
plt.title(state.title())
plt.show()
```
